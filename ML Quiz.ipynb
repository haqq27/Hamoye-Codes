```python
#Q13
#droping columns date and light
new_energy_df = energy_df.drop(columns=["date", "lights"])

#importing minmaxscaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

#normalising our dataset with the minmaxscscaler
normalised_energy_df = pd.DataFrame(scaler.fit_transform(new_energy_df), columns = new_energy_df.columns)

#selecting features 
features_df = normalised_energy_df.drop(columns=['Appliances'])

#selecting our target
target_variable = normalised_energy_df['Appliances']
```

```python
from sklearn.model_selection import train_test_split
#splitting the data into train and test using a (70 - 30)% ratio respectively
x_train, x_test, y_train, y_test = train_test_split(features_df, target_variable, test_size=0.3, random_state= 42)

#importing linear regerssion model
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()

# fitting our train data with the linear regression model
linear_model.fit(x_train, y_train)

#getting our predicted value using the linear regression model
predicted_values = linear_model.predict(x_test)
```

```python
predicted_values
```

```python
#Q13
from sklearn.metrics import mean_absolute_error

#Calculating Mean Absolute Error
mae = mean_absolute_error(y_test, predicted_values)
round(mae, 2)
```

```python
#Q14

#Calculating Residual Sum Squares

rss = np.sum(np.square(y_test - predicted_values))
round(rss, 2)
```

```python
#Q15
from sklearn.metrics import mean_squared_error

#Calculating Root Mean Squared Error
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
round(rmse, 3)
```

```python
#Q16
#Coefficient of determination or R-Squared
from sklearn.metrics import r2_score

#Calculating Coefficient of determination or R-Squared
r2_score = r2_score(y_test, predicted_values)
round(r2_score, 2)
```

```python
#function to get weight or coefficients of a model
def get_weights_df(model, features, col):
    
    weights = pd.Series(model.coef_, features.columns).sort_values()
    
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col]
    
    
    return weights_df
    
```

```python
#Q17
#weights of the linear regression model
linear_weights = get_weights_df(linear_model, x_train, 'Linear Model Weights')
linear_weights
```

```python
#importing Ridge regression model
from sklearn.linear_model import Ridge

#initializing ridge model with alpha = 0.4
ridge_model = Ridge(alpha=0.4)

#fitting our train data with the Ridge regression model
ridge_model.fit(x_train, y_train)

#getting our predicted value using the Ridge regression model
r_values = ridge_model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error

#Calculating Root Mean Squared Error of Ridge regression
r_rmse = np.sqrt(mean_squared_error(y_test, r_values))
round(r_rmse, 3)
```
